```python
X = new_df['name','year','km_driven','fuel_CNG','fuel_Diesel','fuel_Electric','fuel_LPG','fuel_Petrol','seller_type_Dealer','seller_type_Individual','seller_type_Trustmark Dealer','transmission_Automatic','transmission_Manual','owner_First Owner','owner_Fourth & Above Owner','owner_Second Owner','owner_Test Drive Car','owner_Third Owner']
y = new_df['selling_price']
```

```python
#check for multicolinearity, if we are using linear model then we need to check it, it is measured by Variance influence factor
import numpy as np
X = np.array(X)
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif = [variance_inflation_factor(X,i) for i in range(X.shape[1])]
print(vif)
X = pd.DataFrame(X)
```

```python
#Submitted by tgoswami on 03/07/2021 - 13:34. If there is perfect correlation, then VIF = infinity
```

```python
new_df.columns
```

```python
#sampling the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn import linear_model
```

```python
#calling the model 
reg = linear_model.LinearRegression()
```

```python
#training the model
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,color='black')   #your test and pred must be same
```

```python
#Test evaluation
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_test,y_pred)*100) #coefficient of determination
print(mean_squared_error(y_test,y_pred)) #MSE
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = reg, X =X ,\
                             y = y, cv = 15)
print("Accuracy Mean {} Accuracy Variance \ {}".format(accuracies.mean()*100,accuracies.std()*100))
```
